### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

```python
# 元の文字列
original_string = "datascience"

# reversed() で文字を逆順にする。join() で結合して文字列に戻す
reversed_string_2 = "".join(reversed(original_string))

# 結果を出力
print(reversed_string_2)
```

### 問2

「ヘエルスシトンニキア」という文字列の1,3,5,7,9文字目を取り出して連結した文字列を出力してください


```python
# 元の文字列
original_string = "ヘエルスシトンニキア"

# 2つおきに文字を抽出
# [0::2] は、開始位置を0、終了位置を指定せず、2つおきに指定する。
result_string = original_string[0::2]

# 結果を出力
print(result_string)
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
# 元の二つの文字列
string1 = "ヘルシンキ"
string2 = "エストニア"

# zip関数で二つの文字列の文字を先頭から順にペアにする
combined_pairs = zip(string1, string2)

# ペアを一つずつ取り出し、結合してから、すべての結合した文字を一つの文字列にする
# joinで文字列に結合する
result_string = "".join([char for pair in combined_pairs for char in pair])

# 結果を出力
print(result_string)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
# 元の文字列
sentence = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"

# split関数で文字列を単語に分解
words = sentence.split()

# 各単語の文字数を取得し、リストに格納
# lenで単語ごとの文字数を取得する
char_counts = [len(word) for word in words]

# 結果を出力
print(f"単語リスト: {words}")
print(f"文字数リスト: {char_counts}")
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
import re

sentence = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."

# 句読点を取り除き、すべて小文字にしてから単語に分割
words = re.findall(r'\b\w+\b', sentence.lower())
```

```python
# 指定された単語の位置
target_positions = [1, 2, 5, 6, 8, 9, 16]

result_dict = {}

# 単語の位置リストを反復処理
for position in target_positions:
    # Pythonのインデックスに変換
    index = position - 1

    # 単語と先頭2文字を取得
    word = words[index]
    first_two_chars = word[:2]

    # 辞書に追加する
    result_dict[first_two_chars] = position

# 結果を出力
print(result_dict)
```
